```python
import gym
import csv
import random as rnd
import pandas as pd

env = gym.make('LunarLander-v2')

count_of_episodes = 100
count_of_steps = 200
id_global = 0
with open("lander_data_episodes.csv", mode="w", encoding='utf-8') as w_file:
    file_writer = csv.writer(w_file, delimiter = ",", lineterminator='\r')
    file_writer.writerow(["ID", "ID_episode", "X", "Y", "VelX", "VelY", "Angle", "AngularVel", "ActionID",
                         "ResX", "ResY", "ResVelX", "ResVelY", "ResAngle", "ResAngularVel", "TotalReward"])
    cur_episodes = 0
    while cur_episodes < 100:
        observation = env.reset()
        trajectory = []
        reward_for_episode = 0
        for step_ind in range(count_of_steps):
            env.render()
            observation_before = observation
            action = env.action_space.sample()
            id_action = rnd.randint(0,3)
            
            observation, reward, done, info = env.step(id_action)
            
            reward_for_episode += reward
            trajectory.append([id_global, cur_episodes,
                                round(observation_before[0], 2), round(observation_before[1], 2), round(observation_before[2], 2),
                                round(observation_before[3], 2), round(observation_before[4], 2), round(observation_before[5], 2),
                                id_action,
                                round(observation[0], 2), round(observation[1], 2), round(observation[2], 2),
                                round(observation[3], 2), round(observation[4], 2), round(observation[5], 2),
                                reward_for_episode])
            id_global += 1
                
            if done:
                for i in range(len(trajectory)):
                    trajectory[i][15] = trajectory[len(trajectory) - 1][15]
                    file_writer.writerow(trajectory[i])
                print("Episode {} finished after {} timesteps".format(cur_episodes, step_ind + 1))
                cur_episodes += 1
                trajectory = []
                break
env.close()
```
